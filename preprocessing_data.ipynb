```python
import pandas as pd

columns=[ 'age','workclass','fnlwgt','education','education-num','marital-status',
    'occupation','relationship','race','sex','capital-gain','capital-loss',
    'hours-per-week','native-country','income']
df1=pd.read_csv(r'C:\Users\HP\Documents\GitHub\ML_Research\dataset\adult.data', header=None, names=columns)
df2=pd.read_csv(r'C:\Users\HP\Documents\GitHub\ML_Research\dataset\adult.test', header=None, names=columns, skiprows=1)
df2['income'] = df2['income'].str.replace('.', '', regex=False)
df = pd.concat([df1, df2], ignore_index=True)
df.to_csv(r'C:\Users\HP\Documents\GitHub\ML_Research\dataset\adult.csv', index=False)
```

```python
df.replace(' ?', pd.NA, inplace=True)
df.dropna(inplace=True)# Handles missing values
```

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['income'] = le.fit_transform(df['income'])  # 0 or 1
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

num_cols = ['age','fnlwgt','education-num','capital-gain','capital-loss','hours-per-week']
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df.to_csv('adult_clean.csv', index=False)
```
