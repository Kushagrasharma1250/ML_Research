```python
import pandas as pd
import numpy as np
import time
from sklearn.model_selection import train_test_split, cross_validate, StratifiedKFold
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression, Ridge, Lasso, ElasticNet
from sklearn.svm import SVC, LinearSVC
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.neural_network import MLPClassifier, MLPRegressor
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv(r'C:\Users\HP\Documents\GitHub\ML_Research\dataset\adult_clean.csv')
```

```python
X = df.drop('income', axis=1)
y = df['income']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Ridge Regression": Ridge(),
    "Lasso Regression": Lasso(),
    "Elastic Net": ElasticNet(),
    "Linear SVM": LinearSVC(max_iter=10000),
    "SVM RBF": SVC(kernel='rbf', probability=True),
    "SVM Poly": SVC(kernel='poly', probability=True),
    "Decision Tree Classifier": DecisionTreeClassifier(),
    "Decision Tree Regressor": DecisionTreeRegressor(),
    "Random Forest Classifier": RandomForestClassifier(),
    "Random Forest Regressor": RandomForestRegressor(),
    "AdaBoost": AdaBoostClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "Gaussian NB": GaussianNB(),
    "Multinomial NB": MultinomialNB(),
    "Bernoulli NB": BernoulliNB(),
    "k-NN Classifier": KNeighborsClassifier(),
    "k-NN Regressor": KNeighborsRegressor(),
    "MLP Classifier": MLPClassifier(max_iter=1000),
    "MLP Regressor": MLPRegressor(max_iter=1000),
}
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scoring = ['accuracy', 'precision', 'recall', 'f1', 'roc_auc']
```

```python
results = {}
train_times = {}

for name, model in models.items():
    start = time.time()
    try:
        scores = cross_validate(model, X_train, y_train, cv=cv, scoring=scoring)
        results[name] = {metric: np.mean(scores[f'test_{metric}']) for metric in scoring}
    except Exception as e:
        print(f"Error evaluating {name}: {e}")
        results[name] = {metric: None for metric in scoring}
    end = time.time()
    train_times[name] = end - start
```
